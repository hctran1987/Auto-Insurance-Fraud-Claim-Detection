## 3.9 Grid Search for the best model

It seems that SVM, Random forest, Adaboost with random forests, and XGBoost models all perform well. For the simplicity and interpretability I choose the Random forest model to proceed.

```python
rf = RandomForestClassifier(random_state = 42)
params_rf = {'n_estimators': [40, 80, 160, 320, 400, 560, 640, 720]}
grid_rf = GridSearchCV(rf, param_grid = params_rf, cv = 5, scoring = 'roc_auc')
grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_estimator_,grid_rf.best_score_
```

```python
best_model = grid_rf.best_estimator_
```

## 4. Conclusion

It seems that the Random Forest with 560 estimators is the best model above. Now we check the final model performace on the test set.

## 4.1 The AUC scores

```python
best_model.fit(X_train, y_train)
```

```python
y_train_pred_proba = best_model.predict_proba(X_train)
y_test_pred_proba = best_model.predict_proba(X_test)
```

```python
print('The AUC score for the performance on the test set:\n', roc_auc_score(y_test,y_test_pred_proba[:,1]))
```

```python
print('The AUC score for the performance on the train set:\n', roc_auc_score(y_train,y_train_pred_proba[:,1]))
```

## 4.2 The accuracy, precision, recall, f1-scores

```python
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)
```

```python
print('The classification report for the performance on the test set:\n', classification_report(y_test,y_test_pred))
```

```python
print('The classification report for the performance on the train set:\n', classification_report(y_train,y_train_pred))
```
